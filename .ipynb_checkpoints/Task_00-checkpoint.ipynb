```python
import numpy as np
import matplotlib.pyplot as plt
```

# **Signals:**
A signal is an electrical or electromagnetic current that is used for carrying data from one device or network to another. In a data communication system a signal is basically a physical form of data. The physical layer of the OSI model converts the data in the form of signals. Two nodes in a communication system exchange data in the form of signals.

**Different types of Signal plotting**

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(20, 10))
fig.suptitle('Different subplots of a sinusoidal signal', fontsize=15)
time = np.arange(0,1,0.008)
sine_signal = np.sin(2*np.pi*4*time)

ax[0, 0].plot(time,sine_signal)
ax[0, 0].set_title('continuous plot', fontsize = 15)
ax[0,0].axis([0,1,-3,3])
ax[0,0].grid(True)

ax[0, 1].stem(time,sine_signal)
ax[0, 1].set_title('stem plot', fontsize = 15)
ax[0,1].axis([0,1,-3,3])

ax[1,0].scatter(time,sine_signal)
ax[1,0].set_title('scatter plot', fontsize = 15)
ax[-1,-1].axis('off')
ax[1,0].axis([0,1,-3,3])
```

***Continuous time signal and Discrete time signal***

A sine wave is a continuous time signal

A unit step signal is a discrete time signal

```python
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(14, 8))

time = np.arange(0,2,0.005)
sinewave = np.sin(2*np.pi*5*time)

#unit step signal
discontinuous_signal = []
for i in range(-25,25,1):
  if(i>=0):
    discontinuous_signal.append(1)
  else:
    discontinuous_signal.append(0)
positions = np.arange(-25,25,1)

ax[0].plot(time, sinewave)
ax[0].set_title("Continuous Time Signal")
ax[0].axis([0,2,-3,3])
ax[0].grid(True)

ax[1].stem(positions,discontinuous_signal)
ax[1].set_title("Discrete Time Signal")
ax[1].axis([-25,25,-.75,2,])
ax[1].grid(False)
```



***Analog and Digital Signals***

```python
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(14, 8))

time = np.arange(0,2,0.005)
sinewave = np.sin(2*np.pi*5*time)

digital_signal = []
for i in range(20):
  if i%2==0:
    digital_signal.append(0)
  else:
    digital_signal.append(1)


ax[0].plot(time, sinewave)
ax[0].set_title("Analog Signal")
ax[0].axis([0,2,-3,3])
ax[0].grid(True)

ax[1].plot(digital_signal,drawstyle='steps-pre')
ax[1].set_title("Digital Signal")
ax[1].axis([0,20,-.25,1.25])
```

**Sinusoids**

Sinusoidal Signals are periodic functions which are based on the sine or cosine function from trigonometry. The general form of a Sinusoidal Signal is:
$$x(t) = A\sin(2\pi f t + phase)$$
or 
$$x(t) = A\cos(2\pi f t + phase)$$

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(20, 10))

time = np.arange(0,1,0.001)
cosine_signal1 = np.cos(2*np.pi*6*time)
cosine_signal2 = 10*np.cos(2*np.pi*20*time-180)
sine_signal1 = np.sin(2*np.pi*6*time)
sine_signal2 = 10*np.sin(2*np.pi*20*time-180)

fig.suptitle("Different types of sinusoids")
ax[0,0].plot(time,cosine_signal1)
ax[0,0].set_title("cosine signal having zero phase")
ax[0,0].axis([0,1,-3,3])
ax[0,0].grid(True)

ax[0,1].plot(time,cosine_signal2)
ax[0,1].set_title("cosine signal having 180 degree phase and amplitude of 10")
ax[0,1].axis([-0.025,1,-15,15])
ax[0,1].grid(True)

ax[1,0].plot(time,sine_signal1)
ax[1,0].set_title("sine signal having zero phase")
ax[1,0].axis([0,1,-3,3])
ax[1,0].grid(True)

ax[1,1].plot(time,sine_signal2)
ax[1,1].set_title("sine signal having 180 degree phase and amplitude of 10")
ax[1,1].axis([-0.025,1,-15,15])
ax[1,1].grid(True)


```

# **Characteristics of Signals**

**composite signal:**

A composite signal is made of many simple sine waves.

```python
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(16, 8))

time = np.arange(0,1,0.001)
cosine_signal1 = 5*np.cos(2*np.pi*3*time)
cosine_signal2 = 10*np.cos(2*np.pi*6*time)
cosine_signal3 = 15*np.cos(2*np.pi*9*time)

composite = cosine_signal1+cosine_signal2+cosine_signal3
fig.suptitle("Composition of Sinusoids")
ax[0].plot(time,cosine_signal1)
ax[0].plot(time,cosine_signal2)
ax[0].plot(time,cosine_signal3)
ax[0].axis([0,1,-25,25])
ax[0].grid(True)
ax[0].set_title("Three different Sinusoids")

ax[1].plot(time,composite)
ax[1].axis([0,1,-50,50])
ax[1].grid(True)
ax[1].set_title("A Composite Sinusoid")
```

**Shifting and Scaling Property :**

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(20, 10))
fig.suptitle("Shifting and Scaling")

time = np.arange(0,1,0.001)
time_shift = np.arange(-.1,.9,0.001)
sine_signal1 = np.sin(2*np.pi*6*time)
sine_signal2 = np.sin(2*np.pi*10*time+180)
sine_signal3 = 5+sine_signal1


ax[0,0].plot(time,sine_signal1)
ax[0,0].set_title("sine signal having zero phase")
ax[0,0].axis([0,1,-3,3])
ax[0,0].grid(True)

ax[0,1].plot(time_shift,sine_signal2)
ax[0,1].set_title("sine signal left shifted 180 phase")
ax[0,1].axis([-.1,.9,-3,3])
ax[0,1].grid(True)

ax[1,0].plot(time,sine_signal3)
ax[1,0].set_title("sine signal with vertical shift 5 ")
ax[1,0].axis([0,1,-3,8])
ax[1,0].grid(True)

ax[1,1].plot(time,sine_signal1*2)
ax[1,1].set_title("sine signal with  amplitude Scaled by 2 ")
ax[1,1].axis([0,1,-3,3])
ax[1,1].grid(True)
```

***Sampling :***

```python


time = np.arange(0,1,0.0005)
sine_signal1 = np.sin(2*np.pi*6*time)


fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(20, 10))
fig.suptitle("Sampling of Signals")

ax[0,0].plot(time,sine_signal1)
ax[0,0].set_title("Simple sine wave with frequency 6")
ax[0,0].grid(True)
ax[0,0].axis([0,1,-3,3])

sampling_freq = 12
step = 1.0/sampling_freq
sample_positions = np.arange(0,1,step)
sampled_sine_signal = np.sin(2*np.pi*6*sample_positions)

ax[0,1].stem(sample_positions,sampled_sine_signal)
ax[0,1].set_title("With 12HZ sampling rate")

sampling_freq = 24
step = 1.0/sampling_freq
sample_positions = np.arange(0,1,step)
sampled_sine_signal = np.sin(2*np.pi*6*sample_positions)

ax[1,0].stem(sample_positions,sampled_sine_signal)
ax[1,0].set_title("With 24HZ sampling rate")
ax[1,0].axis([0,1,-3,3])

sampling_freq = 48
step = 1.0/sampling_freq
sample_positions = np.arange(0,1,step)
sampled_sine_signal = np.sin(2*np.pi*6*sample_positions)

ax[1,1].stem(sample_positions,sampled_sine_signal)
ax[1,1].set_title("With 48HZ sampling rate")
ax[1,1].axis([0,1,-3,3])
```
