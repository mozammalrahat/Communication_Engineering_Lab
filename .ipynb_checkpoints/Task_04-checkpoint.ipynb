# **ANALOG-TO-ANALOG CONVERSION**
Analog-to-analog conversion, or analog modulation, is the representation of analog
information by an analog signal.
Analog-to-analog conversion can be accomplished in three ways:
1. Frequency modulation (FM)
2. Amplitude modulation (AM)
3. Phase modulation (PM)

# **Frequency Modulation (FM)**
In FM transmission, the frequency of the carrier signal is modulated to follow the chang-
ing voltage level (amplitude) of the modulating signal. The peak amplitude and phase of
the carrier signal remain constant, but as the amplitude of the information signal
changes, the frequency of the carrier changes correspondingly.

```python
#Necessary library
import matplotlib.pylab as plt
import numpy as np

```

```python
message_frequency = 1  
carrier_frequency=15 
modulation_index=10
time = np.arange(0,3,0.001)

modulation_signal=np.sin(2*np.pi*message_frequency*time)  # amplitude 1

plt.figure(figsize=(10,4))
plt.plot(time,modulation_signal)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.axis([0,3,-3,3])
plt.title('Message Signal')
plt.grid(True)
plt.show()
```

```python
carrier_signal = np.sin(2*np.pi*carrier_frequency*time); #carrier wave having amplitude 1

plt.figure(figsize=(10,4))
plt.plot(time,carrier_signal)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.axis([0,3,-3,3])
plt.title('Carrier Signal')
plt.grid(True)
plt.show()
```

```python
fm_signal=np.sin(2*np.pi*carrier_frequency*time+(modulation_index*np.sin(2*np.pi*message_frequency*time)))#Frequency Modulated Signal

plt.figure(figsize=(10,4))
plt.plot(time,fm_signal,'r')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.axis([0,3,-3,3])
plt.title('FM Signal')
plt.grid(True)
plt.show()
```

# **Amplitude Modulation (AM)**
In AM transmission, the carrier signal is modulated so that its amplitude varies with the
changing amplitudes of the modulating signal. The frequency and phase of the carrier
remain the same; only the amplitude changes to follow variations in the information.

```python
message_frequency= 4
carrier_amplitude = 1
carrier_frequency= 20
modulation_index=  10 

message_amplitude = modulation_index * carrier_amplitude
time = np.arange(0,1,0.001)
```

```python
message_signal = np.sin(2*np.pi*message_frequency*time) # message signal having amplitude 1

plt.figure(figsize=(10,4))
plt.plot(time,message_signal)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.axis([0,1,-3,3])
plt.title('Message signal')
plt.grid(True)
plt.show()
```

```python
carrier_signal= carrier_amplitude * np.sin(2*np.pi*carrier_frequency*time) # carrier signal

plt.figure(figsize=(10,4))
plt.plot(time,carrier_signal)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.axis([0,1,-3,3])
plt.title('Carrier signal')
plt.grid(True)
plt.show()
```

```python
am_signal = (1+modulation_index*message_signal) * (carrier_amplitude*carrier_signal)

plt.figure(figsize=(10,4))
plt.plot(time,am_signal)
plt.plot(time,am_signal,'r')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.axis([0,1,-15,15])
plt.title('Modulated signal')
plt.grid(True)
plt.show()
```

# **Phase Modulation (PM)**
In PM transmission, the phase of the carrier signal is modulated to follow the changing
voltage level (amplitude) of the modulating signal.The peak amplitude and frequency of the carrier signal remain constant, but as the amplitude of the information signal
changes, the phase of the carrier changes correspondingly.

```python
message_frequency = 5
carrier_frequency= 100
modulation_index = 20

#all signals have same amplitude of 1

time = np.arange(0, 1,0.001)
y1 = np.cos(2*np.pi*message_frequency*time)
y2 = np.cos(2*np.pi*carrier_frequency*time)
y3 = np.cos((2*np.pi*carrier_frequency*time+(modulation_index*y1)))
```

```python
plt.figure(figsize=(10,4))
plt.plot(time, y1, '-')
plt.grid(True)
plt.axis([0, 1, -3, 3])
plt.title('Phase Modulation')
plt.ylabel('Carrier')
```

```python
plt.figure(figsize=(10,4))
plt.plot(time, y2, '-')
plt.grid(True)
plt.axis([0, 1, -3.1, 3.1])
plt.ylabel('Modulator')
```

```python
plt.figure(figsize=(10,4))
plt.plot(time, y3, 'r')
plt.grid(True)
plt.axis([0, 1, -3, 3])
plt.xlabel('time (s)')
plt.ylabel('Modulated Carrier')
plt.figure(figsize=(200,300))
plt.show()
```
